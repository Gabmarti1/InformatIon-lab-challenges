```python
# here the random library is imported
import random

def function():
    # Loop until a the result c=4 is obtained
    while True:
        # Generate random integers A and B between 1 and 9
        A=random.randint(0,9)
        B=random.randint(0,9)
        # Calculate the product of A and B
        c = A*B
        print(f'A values is = {A}, C values is = {c}')
        #check if the result is 4
        if c ==4:
            # In case the result is correct the succesful message is printed
            print('Success!')
            print(f'A values is = {A}, B values is = {B}')
            break
            #Exit the loop
```

```python
function()
```
